### Extraemos la parte numérica de la columna 'Valor total de la obra' para tener un dato de tipo entero

```python
data_salud['Valor total de la obra USD (millones)'] = data_salud['Valor total de la obra USD (millones)'].apply(extract_numeric_value)
```

```python
# Convertir la columna a cadena, eliminar los puntos y reemplazar las comas por puntos
data_salud['Valor total de la obra USD (millones)'] = data_salud['Valor total de la obra USD (millones)'].astype(str)
data_salud['Valor total de la obra USD (millones)'] = data_salud['Valor total de la obra USD (millones)'].str.replace('.', '').str.replace(',', '.')

# Convertir la columna a tipo float
data_salud['Valor total de la obra USD (millones)'] = pd.to_numeric(data_salud['Valor total de la obra USD (millones)'], errors='coerce')

#Dividimos para tener el total de millones en un número más legible
data_salud['Valor total de la obra USD (millones)'] = (data_salud['Valor total de la obra USD (millones)'] / 1000000).astype(float)
```

### Limpiamos la columna 'Monto comprometido', extraemos la parte numérica, llenamos los valores nulos con '0', transformamos a un tipo de dato float

```python
data_salud['Monto comprometido USD (millones)'] = data_salud['Monto comprometido USD (millones)'].apply(extract_numeric_value)
data_salud['Monto comprometido USD (millones)'] = data_salud['Monto comprometido USD (millones)'].fillna('0')

# Convertir la columna a cadena, eliminar los puntos y reemplazar las comas por puntos
data_salud['Monto comprometido USD (millones)'] = data_salud['Monto comprometido USD (millones)'].astype(str)
data_salud['Monto comprometido USD (millones)'] = data_salud['Monto comprometido USD (millones)'].str.replace('.', '').str.replace(',', '.')

# Convertir la columna a tipo float
data_salud['Monto comprometido USD (millones)'] = pd.to_numeric(data_salud['Monto comprometido USD (millones)'], errors='coerce')

#Dividimos para tener el total de millones en un número más legible
data_salud['Monto comprometido USD (millones)'] = (data_salud['Monto comprometido USD (millones)'] / 1000000).astype(float)
```
